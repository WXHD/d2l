矩阵——空间的扭曲

```python
import torch

A = torch.arange(20, dtype=torch.float32).reshape(5, 4)
B = A.clone()  # 复制 分配新内存
print(A)
print(A.T)  # 转置矩阵（对称矩阵A=A.T）
print(A + B)  # 按位加
print(A * B)  # 按位乘
```

```python
print(A.sum())  # 所有元素求和 总是标量
print(A.sum(axis=1))  # 行求和
print(A.sum(axis=1, keepdims=True))  # 行求和 保持维度不变
print(A / A.sum())  # 广播机制 自动扩展维度以匹配形状
```
